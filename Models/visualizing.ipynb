```python
import pandas as pd 
import numpy as np
from matplotlib.ticker import FormatStrFormatter
import matplotlib.pyplot as plt
```

```python
folder_path = "/content/drive/MyDrive/LSTM_POWER_CLUSTERING/data/fl_data/"
def load_frame(fileName):
  return pd.read_csv("folder_path".format(fileName))
```

```python
def get_filtered_df(df, lag):
  new_df = df.loc[df.lags == lag,["units", "rmse"]]
  new_df = new_df.set_index("units")
  return new_df
```

```python
def get_indiv_df(results):
  sx = get_filtered_df(results, 6)
  et = get_filtered_df(results, 8)
  tw = get_filtered_df(results, 12)
  return sx, et, tw
```

```python
def plot_NLU_vs_RMSE(sx, et, tw):
  plt.rcParams.update({'font.size': 18})
  fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,5),sharex='col', sharey='row')
  plt.subplots_adjust(wspace=.05, hspace=.1)
  ax1.yaxis.set_major_formatter(FormatStrFormatter('%.4f')) 
  ax1.plot(sx)
  ax2.plot(et)
  ax3.plot(tw)
  plt.setp(ax1, xlabel='NLU = 6')
  plt.setp(ax2, xlabel='NLU = 8')
  plt.setp(ax3, xlabel='NLU = 12')
  plt.setp(ax1, ylabel='RMSE')
```
